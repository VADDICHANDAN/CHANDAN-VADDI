13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical
column as followa) If the amount is less than 5 Millions then recode as TypeA.
b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
c) If the amount is greater than 7 Millions then recode as TypeC

```python
def catg(a):
    if a<=5000000:
        return 'TypeA'
    elif (a>5000000)&(a<7000000):
        return 'TypeB'
    elif a>7000000: 
        return 'TypeC'
    else:
        return np.NaN
```

```python
mdf['funding_category']=mdf['raised_amount_usd'].apply(catg) # this will add a new column
mdf
## to convert the raised amount category into funding category instead of creating new column
mdf['raised_amount_usd']=mdf['raised_amount_usd'].apply(catg) # this will edit the existing column
mdf
```
